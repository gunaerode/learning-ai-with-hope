```python
def ageCategory(age):
    if (age < 0):
        print("Invalid age")
    elif (age < 13):
        print("Child")
    elif (age < 60):
        print("Adult")
    elif (age < 75):
        print("Citizen")
    else:
        print("Senior Citizen")

age = int(input("Enter your age: "))
ageCategory(age)
```

```python
# Functions with Return
def ageCategory(age):
    if (age < 0):
        return "Invalid age"
    elif (age < 13):
        return "Child"
    elif (age < 60):
        return "Adult"
    elif (age < 75):
        return "Citizen"
    else:
        return "Senior Citizen"

age = int(input("Enter your age: "))
print(ageCategory(age))
```
