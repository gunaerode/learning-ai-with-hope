```python
def ageCategory(age):
    if (age < 0):
        print("Invalid age")
    elif (age < 13):
        print("Child")
    elif (age < 60):
        print("Adult")
    elif (age < 75):
        print("Citizen")
    else:
        print("Senior Citizen")

age = int(input("Enter your age: "))
ageCategory(age)
```

```python
# Functions with Return
def ageCategory(age):
    if (age < 0):
        return "Invalid age"
    elif (age < 13):
        return "Child"
    elif (age < 60):
        return "Adult"
    elif (age < 75):
        return "Citizen"
    else:
        return "Senior Citizen"

age = int(input("Enter your age: "))
print(ageCategory(age))
```

```python
def calculator(calc_type, num1, num2):
    if (calc_type == "add"):
        return f"{num1 + num2}"
    elif (calc_type == "sub"):
        return f"{num1 - num2}"
    else:
        return "Invalid operation"
```

```python
calc_type = input("Enter operation (add / sub): ")
num1 = int(input("Enter first number: "))
num2 = int(input("Enter second number: "))

result = calculator(calc_type, num1, num2)
print("Result:", result)
```

```python
calc_type = input("Enter operation (add / sub): ")
num1 = int(input("Enter first number: "))
num2 = int(input("Enter second number: "))

result = calculator(calc_type, num1, num2)
print("Result:", result)
```

```python
try:
    calc_type = input("Enter operation (add / sub): ").lower()
    num1 = int(input("Enter first number: "))
    num2 = int(input("Enter second number: "))
    
    result = calculator(calc_type, num1, num2)
    print("Result:", result)
except ValueError as error:
    print(f"Please enter valid integers only {error}")
```
