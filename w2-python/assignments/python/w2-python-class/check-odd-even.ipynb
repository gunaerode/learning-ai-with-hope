```python
# ================= Odd or Even =================
class OddEven:
    @staticmethod
    def check(number):
        if number % 2 == 0:
            print(f"{number} is Even number")
        else:
            print(f"{number} is Odd number")


def odd_even_call():
    try:
        number = int(input("Enter a number: "))
        OddEven.check(number)
    except ValueError:
        print("Invalid number")
```

```python
odd_even_call()
```
