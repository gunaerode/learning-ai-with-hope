```python
# ================= Percentage =================
class FindPercent:
    @staticmethod
    def percentage():
        subjects = {
            "Subject1": 98,
            "Subject2": 87,
            "Subject3": 95,
            "Subject4": 95,
            "Subject5": 93
        }

        total = sum(subjects.values())
        percentage = (total / (len(subjects) * 100)) * 100

        for subject, marks in subjects.items():
            print(f"{subject}= {marks}")

        print(f"Total : {total}")
        print(f"Percentage : {percentage}")


def percentage_call():
    FindPercent.percentage()
```

```python
percentage_call()
```
