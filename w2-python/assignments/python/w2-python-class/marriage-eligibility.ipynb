```python
# ================= Marriage Eligibility =================
class EligibilityForMarriage:
    @staticmethod
    def eligible(gender, age):
        gender = gender.lower()
        if (gender == "male" and age >= 21) or (gender == "female" and age >= 18):
            print("ELIGIBLE")
        else:
            print("NOT ELIGIBLE")


def marriage_eligibility_call():
    try:
        gender = input("Your Gender: ")
        age = int(input("Your Age: "))
        EligibilityForMarriage.eligible(gender, age)
    except ValueError:
        print("Invalid age input")
```

```python
marriage_eligibility_call()
```
