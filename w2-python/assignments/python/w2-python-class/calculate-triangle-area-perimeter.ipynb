```python
# ================= Triangle =================
class Triangle:
    @staticmethod
    def area(height, breadth):
        print("Area formula: (Height * Breadth) / 2")
        print(f"Area of Triangle: {(height * breadth) / 2}")

    @staticmethod
    def perimeter(h1, h2, breadth):
        print("Perimeter formula: Height1 + Height2 + Breadth")
        print(f"Perimeter of Triangle: {h1 + h2 + breadth}")


def triangle_call():
    try:
        print("------ Calculate Area------")
        
        height = float(input("Enter height: "))
        breadth = float(input("Enter breadth: "))
        Triangle.area(height, breadth)

        print("------ Calculate Perimeter------")

        h1 = float(input("Enter height1: "))
        h2 = float(input("Enter height2: "))
        br = float(input("Enter breadth: "))
        Triangle.perimeter(h1, h2, br)
    except ValueError:
        print("Invalid triangle values")
```

```python
triangle_call()
```
