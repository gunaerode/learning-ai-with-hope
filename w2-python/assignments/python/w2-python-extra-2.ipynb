```python
# print Odd Numbers in the list
tupleNums = (20,10,16,19,25,1,276,188)
for num in tupleNums:
    if(num % 2 == 1):
        print(f"{num} is odd")
```

```python
# print Even numbers in the list
for num in tupleNums:
    if(num % 2 == 0):
        print(f"{num} is even")
```
