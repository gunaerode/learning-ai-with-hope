```python
class Calculator:
    @staticmethod
    def add(a, b):
        return a + b

    @staticmethod # works without static as well but it's wrong
    def multiply(a, b):
        return a * b

    def subtract(self, a, b):
        return a - b

    def divide(self, a, b):
        if b == 0:
            return "Error"
        return a / b

calc = Calculator()
print(Calculator.add(2, 3))        # ✅ 5
print(Calculator.multiply(3, 4))   # ✅ 12
print(calc.subtract(5, 2))         # ✅ 3
print(calc.divide(10, 2))          # ✅ 5.0

```
