```python
bmi = int(input("Enter the BMI Index: "))
if (bmi < 18.5):
    print("Underweight")
elif (bmi < 25):
    print("Normal weight")
elif (bmi < 30):
    print("Overweight")
elif (bmi < 35):
    print("Very Overweight")
else:
    print("Obese")
```
