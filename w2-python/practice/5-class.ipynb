```python
class Calculator:
    @staticmethod
    def add(a, b):
        return a + b

    def subtract(self, a, b):
        return a - b

    def multiply(a, b):
        return a * b

    def divide(self, a, b):
        if b == 0:
            return "Error: Division by zero"
        return a / b

print("Multiply: ", Calculator.multiply(3, 4))
```
